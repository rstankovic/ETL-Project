```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#1..#Races CSV
```

```python
csv_file = "../ETL-Project/races.csv"
races2_df = pd.read_csv(csv_file)
races2_df.head()
```

```python
races2_df = races2_df[['raceId', 'name','year','circuitId']].copy()
races2_df.head()
```

```python
races2_trans= races2_df.rename(columns= {"raceId":"race_id",
                                         "name":"Grand_Prix",
                                         "circuitId":"circuit_id"
                                         })
races2_trans.set_index("race_id", inplace =True)
races2_trans.head()
```

```python
#2..#ConstructorResult CSV
```

```python
csv_file = "../ETL-Project/constructorResults.csv"
cons2_result_df = pd.read_csv(csv_file)

cons2_result_df.head()
```

```python
cons2_result_df = cons2_result_df[['constructorId','raceId','points']].copy()
cons2_result_df
```

```python
cons_result_trans = cons2_result_df.rename(columns = {"constructorId":"constructor_id",
                                                      "raceId":"race_id",
                                                      })
cons_result_trans.set_index("constructor_id", inplace=True)
cons_result_trans.head()
```

```python
#3..#Constructors Csv
```

```python
#4..#ConstructorStanding Csv
```

```python
csv_file = "../ETL-Project/constructorStandings.csv"
cons_standing_df = pd.read_csv(csv_file)
cons_standing_df.head(20)
```

```python
cons_standing_df = cons_standing_df[['constructorStandingsId', 'raceId', 'constructorId','points','wins']].copy()
cons_standing_df.head()
```

```python
#5..#LapTimes Csv
```

```python
csv_file = "../ETL-Project/lapTimes.csv"
laptime_df = pd.read_csv(csv_file)
laptime_df.head()
```

```python
laptimes_df = laptime_df[['driverId','lap','milliseconds' ]].copy()
laptimes_df.head()
```

```python
laptimes_trans = laptimes_df.rename(columns ={'driverId':'driver_id',})
laptimes_trans.set_index('driver_id', inplace=True)
laptimes_trans.head()
```

```python
import numpy as np
```

```python
laptimes_bydriver_df = laptimes_trans.groupby('driver_id').agg(np.mean)
laptimes_bydriver_df.head()
```

```python
laptimes_bydriver_df2= laptimes_bydriver_df.sort_values('milliseconds', ascending =True)
laptimes_bydriver_df2.head()
```
